```python
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import pandas as pd

def make_wordmap(z,y,n):
    plt.scatter(z, y)
    for i, txt in enumerate(n):
        plt.annotate(txt, (z[i], y[i]))
        
  
 
example_sent = """This is a sample sentence, 
                  showing off the stop words filtration."""


stop_words = set(stopwords.words('english'))


no_punct = example_sent.translate(str.maketrans('', '', string.punctuation))
word_tokens = word_tokenize(no_punct)
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

windows_size = 1

def n_neighbors(lst, n):
    result = []
    for i in range(len(lst)):
        left = max(0, i - n)
        right = min(len(lst), i + n + 1)
        neighbors = lst[left:right]
        neighbors.remove(lst[i])
        result.append((lst[i],neighbors))
    return result



context = n_neighbors(filtered_sentence, windows_size)



def make_maincontext(lst):
    result = []
    for x,y in lst:
         for word in y:
            result.append((x,word,1))
    return result


def make_n_sampling(lst, n):
    result = []
    for i in range(len(lst)):
        lower = max(0, i-n)
        upper = min(len(lst)-1, i+n)
        sublist = [lst[j] for j in range(len(lst)) if j < lower or j > upper]
        result.append(sublist)
    return result


df_sampling = pd.DataFrame(make_maincontext(context), columns =['Main', 'Context', 'Label'])

'''
def make_negative_sampling(df):
    df_negative = pd.DataFrame(columns=['Main', 'Context', 'Label'])
    for main in set(df["Main"]):
        neg_sampling = df[(df.Main!=main) & (df.Context!=main)].sample(windows_size-1)
        df_negative = pd.concat((df_negative,neg_sampling))
    return df_negative
'''


make_n_sampling(filtered_sentence, window_size)
```
